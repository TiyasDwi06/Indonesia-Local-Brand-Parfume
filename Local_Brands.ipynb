# import library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# read csv files
parfume = pd.read_csv('C:/Users/LENOVO/Documents/Data Science/Data Percobaan 2/Indonesian Local Perfume Brands/Indonesian Local Perfume Brands.csv')
```

```python
parfume.head()
```

```python
# view missing data
parfume.isnull().sum()
```

```python
# view columns
parfume.columns
```

```python
# delete unused columns
parfume.drop(columns=['base_notes3', 'base_notes4',
       'base_notes5', 'base_notes6', 'Unnamed: 31', 'Unnamed: 32',
       'Unnamed: 33', 'base_notes7','mid_notes3', 'mid_notes4',
       'mid_notes5', 'mid_notes6', 'mid_notes7', 'Unnamed: 23', 'Unnamed: 24', 'top_notes3', 'top_notes4', 'top_notes5',
       'top_notes6'], inplace=True)
```

```python
parfume.head()
```

```python
# change the column name
parfume.rename(columns={'top_notes1':'top_1', 'top_notes2':'top_2','mid_notes1':'mid_1','mid_notes2':'mid_2','base_notes1':'base_1','base_notes2':'base_2'}, inplace=True)
```

```python
# Fill in empty data in several columns that contain missing values
parfume['top_2'].mode()
```

```python
parfume['top_2'].fillna('Bergamot', inplace=True)
```

```python
parfume['mid_1'].mode()
```

```python
parfume['mid_1'].fillna('Jasmine', inplace=True)
```

```python
parfume['mid_2'].mode()
```

```python
parfume['mid_2'].fillna('Rose', inplace=True)
```

```python
parfume['base_1'].mode()
```

```python
parfume['base_1'].fillna('Vanilla', inplace=True)
```

```python
parfume['base_2'].mode()
```

```python
parfume['base_2'].fillna('Amber', inplace=True)
```

```python
parfume.isnull().sum()
```

```python
# perform data separation using split
variant = parfume['variant_link'].str.split('/', expand=True)
```

```python
variant.head()
```

```python
variant.rename(columns={2 : 'e_commerce'}, inplace=True)
```

```python
parfume['e_commerce'] = variant['e_commerce']
```

```python
#change the data in the e-commerce column
parfume['e_commerce'] = parfume['e_commerce'].str.replace('.co.id','', regex=True)
parfume['e_commerce'] = parfume['e_commerce'].str.replace('.link','', regex=True)
parfume['e_commerce'] = parfume['e_commerce'].str.replace('.sirclo.me','', regex=True)
parfume['e_commerce'] = parfume['e_commerce'].str.replace('www.','', regex=True)
parfume['e_commerce'] = parfume['e_commerce'].str.replace('.com','', regex=True)
```

```python
parfume['e_commerce'] = parfume['e_commerce'].str.replace('azwaryperfume','tokopedia', regex=True)
```

```python
parfume['e_commerce'].value_counts()
```

```python
parfume.drop(columns=['variant_link','instagram_link'], inplace=True)
```

```python
parfume.head()
```

```python
parfume['price'] = parfume['price'].str.replace(',','')
parfume['price'] = parfume['price'].str.replace('.00','')
parfume['price'] = parfume['price'].astype('int')
```

```python
# Change the data in the price_per_ml column to make calculations easier
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace(',','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.00','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.33','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.27','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.50','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.67','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.57','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.86','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.99','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.64','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.05','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.43','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.71','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.88','')
parfume['price_per_ml'] = parfume['price_per_ml'].str.replace('.48','')
parfume['price_per_ml'] = parfume['price_per_ml'].astype('int')
```

# Visualisasi 

```python
Visualisasi_commerce = parfume['e_commerce'].value_counts().reset_index(name='Counts')
Visualisasi_commerce_plot = sns.barplot(data=Visualisasi_commerce, x='e_commerce', y='Counts')
Visualisasi_commerce_plot.set_xlabel('e-commerce', fontweight='bold')
Visualisasi_commerce_plot.set_ylabel('Counts', fontweight='bold')
Visualisasi_commerce_plot.set_title('Visualisasi e-Commerce', size = 'xx-large', fontweight='heavy')
Visualisasi_commerce_plot.bar_label(Visualisasi_commerce_plot.containers[0])
plt.show()
```

```python
sns.barplot(data= parfume, x='year', y='price') 
plt.title('Harga tertinggi berdasarkan botol', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
sns.barplot(data=parfume, x='year', y='size')
plt.title('Size paling laku', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('size per ML') 
plt.xticks
plt.show()
```

```python
Visualisasi_concentration = parfume['concentration'].value_counts().reset_index(name='Counts')
Visualisasi_concentration_plot = sns.barplot(data=Visualisasi_concentration, x='concentration', y='Counts')
Visualisasi_concentration_plot.set_xlabel('concentration', fontweight='bold')
Visualisasi_concentration_plot.set_ylabel('Counts', fontweight='bold')
Visualisasi_concentration_plot.set_title('Visualisasi concentration', size = 'xx-large', fontweight='heavy')
Visualisasi_concentration_plot.bar_label(Visualisasi_concentration_plot.containers[0])
plt.show()
```
